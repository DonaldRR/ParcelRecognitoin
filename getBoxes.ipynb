```python
img = cv2.imread(TEST_IMG_PATH)
boxes_f = open(TEST_BBOX_PATH, 'r')
```

```python
imgray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
ret,thresh = cv2.threshold(imgray,127,255,0)
im2, contours, hierarchy = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
```

```python
rects = [cv2.minAreaRect(contour) for contour in contours]
boxes_ = [cv2.boxPoints(rect) for rect in rects]
```

```python
boxes = []
boxes_ = np.array(boxes_, dtype=int)
for box in boxes_:
    x_min = min(box[0][0], box[3][0])
    x_max = max(box[1][0], box[2][0])
    y_min = min(box[0][1], box[1][1])
    y_max = max(box[2][1], box[3][1])
    if x_max - x_min <= 0 or y_max - y_min <= 0:
        continue
    print(x_min, x_max, y_min, y_max)
    if pytesseract.image_to_string(img[x_min:x_max, y_min:y_max]) is not u'':
        boxes.append(box)
```

```python
boxes = []
for line in boxes_f.readlines():
    xys = line.split(',')
    boxes.append([int(t) for t in xys])
boxes = np.reshape(newshape=(len(boxes), 4, 2), a=boxes)

n_boxes = len(boxes)
```

```python
def expand_box(pts, ratio):
    x_pixs = ratio*(abs(pts[0][0] - pts[1][0]))
    y_pixs = ratio*(abs(pts[0][1] - pts[3][1]))
    
    pts[0][0] -= x_pixs
    pts[0][1] -= y_pixs
    pts[1][0] += x_pixs
    pts[1][1] -= y_pixs
    pts[2][0] += x_pixs
    pts[2][1] += y_pixs
    pts[3][0] -= x_pixs
    pts[3][1] += y_pixs
    
    return pts
```

```python
for i in range(n_boxes):
    box = expand_box(boxes[i], 20)
    cv2.imwrite('./test_imgs/boxes/{}.jpg'.format(i), four_point_transform(img, box))
```
